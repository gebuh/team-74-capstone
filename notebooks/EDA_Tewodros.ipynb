```python
loan_amt_businessType = df.groupby('BusinessType').agg(LOAN_AMOUNT =('InitialApprovalAmount','sum'), PAYROLL_PROCEED  =('PAYROLL_PROCEED','sum'),HELATH_PROCEED  =('HEALTH_CARE_PROCEED','sum'))
colors_list = ['#5cb85c','#5bc0de','#d9534f']
loan_amt_businessType.plot(kind='barh', title= 'Approval_Amount vs PAYROLL_Proceed vs Health_Proceed', figsize=(12,8), width=0.8, color=colors_list, edgecolor=None)
```

```python
# Bar plot
loan_amt_businessType = df.groupby('BusinessType').agg(INITIAL_LOAN_AMOUNT =('InitialApprovalAmount','sum'), RENT_PROCEED  =('RENT_PROCEED','sum'),PAYROLL_PROCEED  =('PAYROLL_PROCEED','sum'),HELATH_PROCEED  =('HEALTH_CARE_PROCEED','sum'))
loan_amt_businessType.plot(kind='barh', title= 'Approval_Amount vs Payroll_ Proceed Vs Rent_Proceed vs Health Proceed',cmap='Dark2', figsize=(15,10), rot=40);
```

```python
loan_amt_businessType = df.groupby('BusinessType').agg(Current_Approval_Amount =('CurrentApprovalAmount','sum'), RENT_PROCEED  =('RENT_PROCEED','sum'),PAYROLL_PROCEED  =('PAYROLL_PROCEED','sum'),HELATH_PROCEED  =('HEALTH_CARE_PROCEED','sum'))
loan_amt_businessType.plot(kind='barh', title= 'Approval_Amount vs Payroll_ Proceed Vs Rent_Proceed vs Health Proceed',cmap='Dark2', figsize=(15,10), rot=40);
```

```python
# group the data
loan_amt_businessType = df.groupby('BusinessType').agg(approval_loanAmount =('InitialApprovalAmount','sum'), RENT_PROCEED  =('RENT_PROCEED','sum'))
loan_amt_businessType.plot(kind='line', title= 'Approval_Amount vs Rent_Proceed',cmap='Dark2', figsize=(15,8), rot=40);
```
