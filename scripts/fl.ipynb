```python
base_url = "http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/a/Page1?searchNameOrder=A"

from selenium.common.exceptions import NoSuchElementException        


def scrape(url):
  # is_next = True
  results = []
  # count = 0
  for i in range(4):
    # count += 1
    driver.get(url)
    companies = driver.find_elements_by_class_name('large-width')
    companies_status = driver.find_elements_by_class_name('small-width')

    # results = []
    for i in range(len(companies)-1):
      temp_data = {'Corporate Name': companies[i+1].text,
                    'Status': companies_status[i+1].text}
      results.append(temp_data)
    url = driver.find_element_by_xpath(u'//a[text()="Next List"]').get_attribute('href')
    # if driver.find_element_by_xpath(u'//a[text()="Next List"]').is_displayed():
    #   is_next = True
    #   url = driver.find_element_by_xpath(u'//a[text()="Next List"]').get_attribute('href')
    # if count == 3:
    #   is_next = False
    # else:
    #   is_next = False
  return results

df = pd.DataFrame(scrape(base_url))
```
