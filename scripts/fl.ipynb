#Method 1: For scraping the first page

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
import pandas as pd
```

```python
letter = input("Enter a letter: ")
uc = letter.upper()
lc = letter.lower()
```

```python
base_url = f"http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/{lc}/Page1?searchNameOrder={uc}"
r = requests.get(base_url)
webpage = bs(r.content, features='lxml')
```

```python
table = webpage.find('table')
columns = table.find_all('th')
column_names = [c.string for c in columns]
table_rows = table.find_all('tr')
```

```python
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.string for tr in td ]
    l.append(row)
df = pd.DataFrame(l,columns=column_names)
```

```python
import pandas as pd
!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=chrome_options)
driver.get("http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/a/Page1?searchNameOrder=A")
driver.maximize_window()
```

```python
companies = driver.find_elements_by_class_name('large-width')
companies_status = driver.find_elements_by_class_name('small-width')
```

```python
# print(companies)
for company in companies:
  print(company.text)
```

```python
results = []
for i in range(len(companies)-1):
  temp_data = {'Corporate Name': companies[i+1].text,
                'Status': companies_status[i+1].text}
  results.append(temp_data)

df = pd.DataFrame(results)
```
