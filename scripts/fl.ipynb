###Finding the Pattern

COLE SCOTT & KISSANE PA
http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/COLE%20SCOTT%20%26%20KISSANE%20PA/Page1?searchNameOrder=COLESCOTTKISSANEPA

CITRUS WORLD INC.
http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/CITRUS%20WORLD%20INC./Page1?searchNameOrder=CITRUSWORLD

SNI COMPANIES, INC.
http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/SNI%20COMPANIES%2c%20INC./Page1?searchNameOrder=SNICOMPANIES



```python
def result_link_generator(company_name):
  first_half = quote(company_name)

  terms = prepare_terms()
  # Takes out the legal suffix such as LLC and INC.
  second_half = basename(company_name, terms, prefix=False, middle=False, suffix=True)
  second_half = re.sub(r'\W+', '', second_half)
  return f'http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/{first_half}/Page1?searchNameOrder={second_half}'
```
