```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
import pandas as pd
```

```python
letter = input("Enter a letter: ")
uc = letter.upper()
lc = letter.lower()
```

```python
base_url = f"http://search.sunbiz.org/Inquiry/CorporationSearch/SearchResults/EntityName/{lc}/Page1?searchNameOrder={uc}"
r = requests.get(base_url)
webpage = bs(r.content, features='lxml')
```

```python
table = webpage.find('table')
columns = table.find_all('th')
column_names = [c.string for c in columns]
table_rows = table.find_all('tr')
```

```python
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.string for tr in td ]
    l.append(row)
df = pd.DataFrame(l,columns=column_names)
```
